```python
import cv2, dlib
import numpy as np
from imutils import face_utils
from keras.models import load_model
import time
import threading

IMG_SIZE = (34, 26)
count = ''
cnt=cnt2=0
point = 30,30+(40)
start=time.time()

textTime = 0
isBlink = False #눈을 감아야하는지를 알려주게끔 하는 프레임

#눈 비교 해서 알림 설정 (알림은 3초뒤 사라지게.)

def init_cnt():#cnt 초기화
    global cnt
    cnt=0
    threading.Timer(10 ,init_cnt).start() #숫자가 초,10초마다 cnt를 0으로 초기화    
    
def init_cnt2():#비교 
    global cnt
    global cnt2
    global isBlink
    cnt2=cnt
    if cnt2<=600 and cnt2 >=1: #여기서 눈 깜빡임을 얼마나 추천할 것인지 지금은 600으로 되어있음.
            isBlink = True
            cnt2 = 0 # 9.99초마다 cnt2를 초기화
    threading.Timer(9.999, init_cnt2).start() # 9.99초마다 이 함수를 초기화
    
def cnt2To0():#아마 무쓸모 일단은 넣어놈
    global cnt2
    cnt2=0
    threading.Timer(12, cnt2To0).start()
    
    
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')

model = load_model('models/2022_02_07_00_38_26.h5')
model.summary()

def crop_eye(img, eye_points):
    x1, y1 = np.amin(eye_points, axis=0)
    x2, y2 = np.amax(eye_points, axis=0)
    cx, cy = (x1 + x2) / 2, (y1 + y2) / 2

    w = (x2 - x1) * 1.2
    h = w * IMG_SIZE[1] / IMG_SIZE[0]

    margin_x, margin_y = w / 2, h / 2

    min_x, min_y = int(cx - margin_x), int(cy - margin_y)
    max_x, max_y = int(cx + margin_x), int(cy + margin_y)

    eye_rect = np.rint([min_x, min_y, max_x, max_y]).astype(np.int)

    eye_img = gray[eye_rect[1]:eye_rect[3], eye_rect[0]:eye_rect[2]]

    return eye_img, eye_rect

# main
cap = cv2.VideoCapture('videos/1.mp4')

init_cnt()
init_cnt2()
#cnt2To0()

while True:
    
    ret, img_ori = cap.read()
    
    if not ret:
        break

    img_ori = cv2.resize(img_ori, dsize=(0, 0), fx=0.5, fy=0.5)

    img = img_ori.copy()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    faces = detector(gray)
        
    for face in faces:
        shapes = predictor(gray, face)
        shapes = face_utils.shape_to_np(shapes)

        eye_img_l, eye_rect_l = crop_eye(gray, eye_points=shapes[36:42])
        eye_img_r, eye_rect_r = crop_eye(gray, eye_points=shapes[42:48])

        eye_img_l = cv2.resize(eye_img_l, dsize=IMG_SIZE)
        eye_img_r = cv2.resize(eye_img_r, dsize=IMG_SIZE)
        eye_img_r = cv2.flip(eye_img_r, flipCode=1)

        cv2.imshow('l', eye_img_l)
        cv2.imshow('r', eye_img_r)

        eye_input_l = eye_img_l.copy().reshape((1, IMG_SIZE[1], IMG_SIZE[0], 1)).astype(np.float32) / 255.
        eye_input_r = eye_img_r.copy().reshape((1, IMG_SIZE[1], IMG_SIZE[0], 1)).astype(np.float32) / 255.

        pred_l = model.predict(eye_input_l)
        pred_r = model.predict(eye_input_r)

    # visualize
        state_l = 'O' if pred_l > 0.1 else '-'
        state_r = 'O' if pred_r > 0.1 else '-'

        state_l = state_l % pred_l
        state_r = state_r % pred_r
        

        #cv2.rectangle(img, pt1=tuple(eye_rect_l[0:2]), pt2=tuple(eye_rect_l[2:4]), color=(255,255,255), thickness=2) 이건 눈이 어딘지
        #cv2.rectangle(img, pt1=tuple(eye_rect_r[0:2]), pt2=tuple(eye_rect_r[2:4]), color=(255,255,255), thickness=2) 이건 눈이 어딘지

        #cv2.putText(img, state_l, tuple(eye_rect_l[0:2]), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255,255,255), 2) 이것은 눈이 얼마나 감았는지
        #cv2.putText(img, state_r, tuple(eye_rect_r[0:2]), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255,255,255), 2) 이것은 눈이 얼마나 감았는지
       
        if state_l=='-' or state_r=='-': #눈 감았을떄, 계속 카운트가 올라감 - ->0으로 변할떄만 카운트 하게끔
            cnt+=1
       
        count=str(cnt)
        cv2.putText(img,count,point ,cv2.FONT_HERSHEY_SIMPLEX,0.7,(255,255,255), 2) #눈 깜빡임. 입력값 
        
        #한번 루프가 1프레임이라고 가정시 1초에 60프레임이라고 가정하고 만든 코드임
        if(textTime <= 45 and isBlink): #texttime 는 1프레임  즉 45프레임이 지나면 없어지게 됨. 만약 영상이 느리게 된다면 2초 또는 3초로 늘어날수도 있습니다....
            cv2.putText(img,"Blinking!!",(100,100) ,cv2.FONT_HERSHEY_SIMPLEX,0.7,(0,0,255), 2)
            textTime += 1 #프레임 +1
        else:
            isBlink = False
            textTime = 0

    cv2.imshow('result', img)
    if cv2.waitKey(1) == ord('q'):
        break
```
